```python
for i in range(len(weather_data)):
    date_list_form = weather_data.loc[i, 'datetime'].split('/')
    for j in range(len(date_list_form)):
        if len(date_list_form[j]) == 1:
            date_list_form[j] = '0' + date_list_form[j]
        if j == 2:
            date_list_form[j] = '20' + date_list_form[j]
    date = date_list_form[2] + date_list_form[0] + date_list_form[1]
    date = int(date)
    weather_data.loc[i, 'datetime'] = date
weather_data
```

```python
for i in range(len(NASDAQ_data)):
    date_list_form = NASDAQ_data.loc[i, 'Date'].split('/')
    for j in range(len(date_list_form)):
        if len(date_list_form[j]) == 1:
            date_list_form[j] = '0' + date_list_form[j]
    date = date_list_form[2] + date_list_form[0] + date_list_form[1]
    date = int(date)
    NASDAQ_data.loc[i, 'Date'] = date
NASDAQ_data
```
